```python
import requests
from bs4 import BeautifulSoup
import re

webscrapeDef = requests.get('https://www.webharvy.com/articles/what-is-web-scraping.html')

soup = BeautifulSoup(webscrapeDef.text,'html.parser')
soup
regex = r'\b\w+\b'
scrapeDef = re.findall(regex,soup.findAll('p')[1].text)
"".join(scrapeDef)
```

```python
yellowpages = requests.get('https://www.yellowpages.com/los-angeles-ca/auto-repair')
yellowpages
```

```python
yellowpagesSoup = BeautifulSoup(yellowpages.text,"html.parser")
print(yellowpagesSoup.prettify())
```

```python
# print(yellowpagesSoup.findAll('a'))
autorepair_list = []
for autorepair in yellowpagesSoup.findAll('a',{'class':'business-name'})[0:20]:
    autorepair_list.append(autorepair.text)

x = "".join(autorepair_list)
print(x)
```
