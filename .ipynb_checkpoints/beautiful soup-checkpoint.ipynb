```python
import requests
from bs4 import BeautifulSoup
import re

webscrapeDef = requests.get('https://www.webharvy.com/articles/what-is-web-scraping.html')

soup = BeautifulSoup(webscrapeDef.text,'html.parser')
soup
regex = r'\b\w+\b'
scrapeDef = re.findall(regex,soup.findAll('p')[1].text)
"".join(scrapeDef)
```

```python
trivago = requests.get('https://www.trivago.com/?aDateRange%5Barr%5D=2017-10-09&aDateRange%5Bdep%5D=2017-10-10&aPriceRange%5Bfrom%5D=0&aPriceRange%5Bto%5D=0&iPathId=34470&iGeoDistanceItem=191681&aCategoryRange=0%2C1%2C2%2C3%2C4%2C5&aOverallLiking=1%2C2%2C3%2C4%2C5&bTopDealsOnly=false&iRoomType=7&cpt=19168102&iIncludeAll=0&iViewType=0&bIsSeoPage=false&bIsSitemap=false&')
trivago
```

```python
trivagoSoup = BeautifulSoup(trivago.text,"html.parser")
print(trivagoSoup.prettify())
```

```python
print(trivagoSoup.findAll('h3'))
hotel_list = []
for hotel in trivagoSoup.find_all('h3',{'class':'name__copytext'}):
    hotel_list.append(hotel.text)
    
print(hotel_list)
```
